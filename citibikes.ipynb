```python
# import dependencies
import pandas as pd
```

```python
# load files
feb_2021 = pd.read_csv("Resources/JC-202102-citibike-tripdata.csv")
mar_2021 = pd.read_csv("Resources/JC-202103-citibike-tripdata.csv")
apr_2021 = pd.read_csv("Resources/JC-202104-citibike-tripdata.csv")
may_2021 = pd.read_csv("Resources/JC-202105-citibike-tripdata.csv")
jun_2021 = pd.read_csv("Resources/JC-202106-citibike-tripdata.csv")
jul_2021 = pd.read_csv("Resources/JC-202107-citibike-tripdata.csv")
aug_2021 = pd.read_csv("Resources/JC-202108-citibike-tripdata.csv")
sep_2021 = pd.read_csv("Resources/JC-202109-citibike-tripdata.csv")
oct_2021 = pd.read_csv("Resources/JC-202110-citibike-tripdata.csv")
nov_2021 = pd.read_csv("Resources/JC-202111-citibike-tripdata.csv")
dec_2021 = pd.read_csv("Resources/JC-202112-citibike-tripdata.csv")
jan_2022 = pd.read_csv("Resources/JC-202201-citibike-tripdata.csv")
```

```python
# merge and export
# (https://pandas.pydata.org/docs/reference/api/pandas.concat.html)
merged_frame = [feb_2021, mar_2021, apr_2021, may_2021, jun_2021, jul_2021, aug_2021, sep_2021, oct_2021, nov_2021, dec_2021, jan_2022]
merged_df = pd.concat(merged_frame, ignore_index=True).drop_duplicates()
merged_df.head()
```

```python
merged_df.to_csv("Resources/2021-2022_merged-citibike-tripdata.csv", index=False)
```
